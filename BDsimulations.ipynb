```python
import birdepy as bd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import numpy as np
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
## for Palatino and other serif fonts use:
#rc('font',**{'family':'serif','serif':['Palatino']})
rc('text', usetex=True)

plt.rcParams.update({'font.size': 20})

from matplotlib.offsetbox import AnchoredText
```

```python
# Definition of the transition rates
def custom_b_rate(z, p): return (2*z+1)*(2*z+1+p[0])
def custom_d_rate(z, p): return 2*z*(2*z+1+p[0])
```

```python
def plot_process(alpha,population_size, t_max,col):
    obs_times = [t/t_max for t in range(t_max)]
    pop_sizes = bd.simulate.discrete([alpha], model='custom', z0=population_size, times=obs_times, b_rate=custom_b_rate, d_rate=custom_d_rate, seed=2021)
    plt.step(obs_times, pop_sizes, 'r', where='post', color=col, linewidth = 1,  label=fr'$\alpha={alpha}$')
    plt.ylabel(r'$P(t)$')
    plt.xlabel(r'$t$')
```

```python
plt.rcParams['figure.figsize'] = [9, 6]
plot_process(0.1,100,1000,'tab:purple')
plot_process(5,100,1000,'tab:blue')
plot_process(15,100,1000,'tab:green')
plt.title(f'Initial population 100')
plt.legend(loc="upper left")
```

```python
plt.rcParams['figure.figsize'] = [9, 6]
plot_process(0.1,200,1000,'tab:purple')
plot_process(5,200,1000,'tab:blue')
plot_process(15,200,1000,'tab:green')
plt.title(f'Initial population 200')
plt.legend(loc="upper left")
```
